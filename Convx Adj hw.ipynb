```python
settle_dt = date(2025, 3, 20)
i_start = date(2026, 6, 17)
i_end = date(2026, 9, 16)
```

Normal vol estimated as 101.4 bps 

```python
p = 96.48434697
r = 100 - p
T1 = (i_start - settle_dt).days / 360  
T2 = (i_end - settle_dt).days / 360
mean_rev = 0.03
sigma = 0.001014 # normal vol 
```

```python
b_t1_t2 = (1 - np.exp(-mean_rev * (T2 - T1))) / mean_rev
b_sqrd_0_t1 = ((1 - np.exp(-mean_rev * (T1 - 0))) / mean_rev) ** 2
x_t1_t2 = (sigma ** 2 / (2 * mean_rev)) * b_t1_t2 * (
    b_t1_t2 * (1 - np.exp(-2 * mean_rev * T1)) + (mean_rev * b_sqrd_0_t1)
) * 100
```

```python
convexity = -(1 - np.exp(-x_t1_t2)) * (r / 100 + 1 / (T2 - T1)) * 100
forward = round(r + convexity, 4)
```

```python
print("Convexity:", convexity)
print("Forward:", forward)
```
